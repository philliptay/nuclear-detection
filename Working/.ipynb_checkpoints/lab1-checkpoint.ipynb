```python
# Lab report 1 data
import numpy as np
import matplotlib.pyplot as plt

background = np.loadtxt('background_0309.csv',delimiter=',')

# calibrated with Cs-137 and Co-60
spectrum1_raw = np.loadtxt('calibration.csv',delimiter=',')
spectrum1 = spectrum1_raw - background;

# number of channels
n = np.size(background)
channels = np.arange(0,n)

# export spectrum data
data = np.vstack((channels,spectrum1))
data = data.T
np.savetxt('export.csv',data,delimiter=',')

# DATA VISUALIZATION
# background data
plt.figure(1)
plt.plot(channels,background)
plt.xticks(np.arange(0,n+1,n/8))
plt.title('Background Gamma-ray Spectrum')
plt.xlabel('Channel')
plt.ylabel('Count')

plt.show()
```

```python
# calibrated spectrum
plt.figure(2)
plt.plot(channels,spectrum1)
plt.xticks(np.arange(0,n+1,n/8))
plt.title('Cs-137 and Co-60 Calibration')
plt.xlabel('Channel')
plt.ylabel('Count')

plt.show()
```
